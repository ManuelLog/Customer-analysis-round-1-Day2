```python
print("Maximum",lo["total claim amount"].max())
print("Minimun",lo["total claim amount"].min())
print("Mean",lo["total claim amount"].mean())
```

```python
def tca_bins(x):
    if x <=400:
        return 1
    elif x <=800:
        return 2
    elif x <=1200:
        return 3
    elif x <=1600:
        return 4
    else:
        return 5

```

```python
lo["total claim amount bins"] = list(map(tca_bins, lo["total claim amount"]))
lo
```

```python
lo.groupby("total claim amount bins")["response_rate"].mean().plot.bar()
plt.show()
```
